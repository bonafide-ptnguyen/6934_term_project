### 5. "How" Analysis: Design of Visuals

I create three distinct visualization designs based on a What-Why-How analysis framework.

**(Design 1) Multi-Platform Performance Bar Chart**

*   **Connection to What/Why Target Pairs:**
    *   **What:** Compares song performance metrics across different music platforms (Spotify, YouTube, TikTok, etc.).
    *   **Why:** Addresses **Research Question 1**: To **Compare Distribution** of song performance across **Platforms**.

*   **Idioms Involved:**
    *   **Grouped Bar Chart:** Used to visualize and compare performance metrics for each song across multiple platforms simultaneously.

*   **Methods Used:**
    *   **Normalization:** Normalize performance metrics (e.g., stream counts, playlist reach) across platforms to allow for direct comparison despite different scales.
    *   **Platform Grouping:** Group bars by song, with sub-bars within each group representing different platforms.
    *   **Color Encoding:** Use distinct colors to represent different platforms, making it easy to visually distinguish platform performance for each song.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'datasets/Most_Streamed_Spotify_Songs_2024.csv'
spotify_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Data Cleaning - Convert stream/count columns to numeric, handling commas and errors
columns_to_numeric = [
    'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach',
    'YouTube Views', 'YouTube Likes', 'TikTok Posts', 'TikTok Likes', 'TikTok Views',
    'YouTube Playlist Reach', 'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
    'Deezer Playlist Count', 'Deezer Playlist Reach', 'Amazon Playlist Count',
    'Pandora Streams', 'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts'
]

for col in columns_to_numeric:
    spotify_data[col] = spotify_data[col].astype(str).str.replace(',', '').astype(float)

# Convert 'Explicit Track' to numeric (0 and 1)
spotify_data['Explicit Track'] = pd.to_numeric(spotify_data['Explicit Track'], errors='coerce').fillna(0).astype(int)

print("Data loaded and cleaned successfully!")
```

```python
# Visual Design 1: Multi-Platform Performance Bar Chart

# Select platforms and metrics for comparison
platforms = ['Spotify Streams', 'YouTube Views', 'TikTok Views']
platform_names = ['Spotify', 'YouTube', 'TikTok'] # for better labels
top_n_songs = 10  # Display top N songs for clarity

# Normalize metrics
normalized_data = pd.DataFrame()
for index, row in spotify_data.head(top_n_songs).iterrows():
    song_name = row['Track']
    song_metrics = {}
    max_val = 0
    for platform_col in platforms:
        value = row[platform_col]
        if not pd.isna(value): # Handle NaN values if any
            max_val = max(max_val, value)

    for i, platform_col in enumerate(platforms):
        value = row[platform_col]
        normalized_value = (value / max_val) if max_val > 0 and not pd.isna(value) else 0
        song_metrics[platform_names[i]] = normalized_value
    normalized_data = pd.concat([normalized_data, pd.DataFrame([song_metrics], index=[song_name])])

# Plot the normalized data
ax = normalized_data.plot(kind='bar', figsize=(12, 6))
plt.title('Normalized Performance Across Platforms (Top {} Songs)'.format(top_n_songs))
plt.ylabel('Normalized Performance Score (0-1)')
plt.xlabel('Song Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**(Design 2) Correlation Heatmap of Spotify Popularity Factors**

*   **Connection to What/Why Target Pairs:**
    *   **What:**  Examines the correlation between `Spotify Popularity` and other numerical song attributes.
    *   **Why:** Addresses **Research Question 2**: To **Find Correlation** and explore the **Dependency** between `Spotify Popularity` and other factors.

*   **Idioms Involved:**
    *   **Heatmap:**  Used to display a correlation matrix, visually encoding correlation strength and direction using color intensity and hue.

*   **Methods Used:**
    *   **Correlation Calculation:** Calculate the correlation matrix between `Spotify Popularity` and relevant numerical attributes (e.g., streams, playlist counts, social media engagement).
    *   **Color Scale for Correlation:** Use a diverging color scale (e.g., red-white-blue) to represent correlation coefficients, with red for positive, blue for negative, and intensity indicating strength.
    *   **Attribute Ordering:** Order attributes in the heatmap to potentially reveal clusters or patterns in correlations.

```python
# Visual Design 2: Correlation Heatmap of Spotify Popularity Factors

# Select numerical columns for correlation analysis
correlation_columns = [
    'Spotify Popularity', 'Spotify Streams', 'Spotify Playlist Count', 'Spotify Playlist Reach',
    'YouTube Views', 'YouTube Likes', 'TikTok Views', 'TikTok Likes', 'TikTok Posts',
    'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
    'Deezer Playlist Count', 'Deezer Playlist Reach', 'Amazon Playlist Count',
    'Pandora Streams', 'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts'
]

correlation_data = spotify_data[correlation_columns].copy()

# Calculate correlation matrix
correlation_matrix = correlation_data.corr(numeric_only=True) # Explicitly set numeric_only

# Create heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f") # Removed annot for better visual clarity, can add annot=True for values
plt.title('Correlation Heatmap of Spotify Popularity Factors')
plt.tight_layout()
plt.show()
```

**(Design 3) Explicit Content Impact Box Plots**

*   **Connection to What/Why Target Pairs:**
    *   **What:** Compares the distribution of streaming and engagement metrics for songs with and without explicit content.
    *   **Why:** Addresses **Research Question 3**: To **Compare Distribution** of performance metrics based on `Explicit Track` category.

*   **Idioms Involved:**
    *   **Box Plots:** Used to visualize and compare the distribution of metrics (streams, views, playlist reach) for explicit and non-explicit song categories.

*   **Methods Used:**
    *   **Categorical Grouping:** Group songs into "Explicit" and "Non-Explicit" categories based on the `Explicit Track` attribute.
    *   **Box Plot per Metric:** Create a series of box plots, one for each performance metric (e.g., Spotify Streams, YouTube Views).
    *   **Side-by-Side Comparison:**  Place box plots for "Explicit" and "Non-Explicit" songs side-by-side for each metric to facilitate visual comparison of distributions.

```python
# Visual Design 3: Explicit Content Impact Box Plots

# Select metrics for box plot comparison
boxplot_metrics = [
    'Spotify Streams', 'YouTube Views', 'TikTok Views', 'Spotify Playlist Reach'
]
metric_names = ['Spotify Streams (Millions)', 'YouTube Views (Millions)', 'TikTok Views (Billions)', 'Spotify Playlist Reach (Millions)'] # For better labels and scaling

# Create box plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten() # Flatten axes array for easy iteration

for i, metric in enumerate(boxplot_metrics):
    ax = axes[i]

    # Scale down large numbers for better readability on the plot
    if metric in ['Spotify Streams', 'Spotify Playlist Reach', 'YouTube Views']:
        spotify_data['Scaled_Metric'] = spotify_data[metric] / 1e6  # Scale to millions
    elif metric == 'TikTok Views':
        spotify_data['Scaled_Metric'] = spotify_data[metric] / 1e9  # Scale to billions
    else:
        spotify_data['Scaled_Metric'] = spotify_data[metric] # No scaling for others if needed

    sns.boxplot(y='Scaled_Metric', x='Explicit Track', hue='Explicit Track', data=spotify_data, ax=ax, palette=['lightgreen', 'lightcoral'], dodge=False) # Vertical boxplots, hue='Explicit Track', dodge=False to remove margin
    ax.set_title('{} Distribution by Explicit Content'.format(metric_names[i]))
    ax.set_xlabel('Explicit Track (0: No, 1: Yes)') # x-axis label is Explicit Track
    ax.set_ylabel(metric_names[i]) # y-axis label is the metric

    spotify_data.drop(columns=['Scaled_Metric'], inplace=True) # Clean up temporary scaled column

plt.tight_layout()
plt.show()
```
